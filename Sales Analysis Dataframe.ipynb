#### Convert Month Series to int Dtype

```python
df["Month"] = df["Month"].astype("int32")
```

#### Add a Sales Column 

```python
df["Sales"] = df["Qty"] * df["Price"]
df.head()
```

#### Best month for Sale 

```python
monthlySales = df.groupby("Month").sum()
monthlySales
```

```python
import matplotlib.pyplot as plt

months = range(1, 10)
plt.bar(months, monthlySales["Sales"])
plt.xticks()
plt.xlabel("Months")
plt.ylabel("Sales in USD ($)")
plt.title("Monthly Sales for the Year Ending 2019")
plt.style.use("ggplot")
plt.show()
```

#### City with the highest Sales

#### Alternatively using Apply Function extracts the city from the purchase address

#### Task 3: Group by city to get the highest sale per city

```python
salesPerCity = df.groupby("City").sum()
salesPerCity
```

```python
df["City"].is_unique().head()
```

```python
import matplotlib.pyplot as plt

cities = salesPerCity["City"].unique()
plt.bar(cities, salesPerCity["Sales"])
plt.xticks(cities)
plt.xlabel("State / City")
plt.ylabel("Sales in USD ($)")
plt.title("Sales Per City")
plt.show()
```
