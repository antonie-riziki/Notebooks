#### Open Csv File and Analyse the data
#### May include alot of Data Cleaning

```python
df = pd.read_csv("D:\Open Classroom\Data Files\Pandas-Data-Science-Tasks-master\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data\Annual_Sales.csv")
df.head()
```

### Data Cleaning

```python
df.dropna(axis = "index", how = "any", inplace = True)
```

```python
df.rename(columns = {"Order ID": "Order_ID",
                    "Quantity Ordered": "Qty",
                    "Price Each": "Price",
                    "Order Date": "Date_Ordered",
                    "Purchase Address": "Purchase_Address"}, inplace = True)
```

```python
df.head()
```

```python
df = df[df["Date_Ordered"].str[0:2] != "Or"]
df.head()
```

#### Create a series to hold the month for each purchase 

```python
df["Month"] = df["Date_Ordered"].str[0:2]
df.head()
```

#### Assign Datatype for each series respectively

##### Approach 1 : Using astype function

```python
df["Month"] = df["Month"].astype("int64")
df.head()
```

##### Approach 2 : Using to_numeric pandas module

```python
df["Qty"] = pd.to_numeric(df["Qty"])
df["Price"] = pd.to_numeric(df["Price"])
df.head()
```

### Task 1: Augment the DataFrame with Column Series
#### Sales Column

```python
df["Sales"] = df["Qty"] * df["Price"]
df.head()
```

##### Assigment 1: Find the month with the highest Sales

```python
sales_per_month = df.groupby("Month").sum()
sales_per_month.head(20)
```

#### Create a visualization for the highest sales per month

```python
month = range(1, 13)
plt.bar(month, sales_per_month["Sales"])
plt.xticks(month)
plt.xlabel("Months")
plt.ylabel("Sales in USD ($)")
plt.legend("Sales")
plt.title("Sales for the Business Year 2019")
plt.style.use("ggplot")
plt.show()
```

##### Assignment 2: Find the highest sales per city

#### City Column

##### Approach 1 : Using pandas str method to split function

```python
df[["Add_1", "City", "Add_2"]] = df["Purchase_Address"].str.split(",", expand = True)
df.head()
```

##### drop unnecessary columns

```python
df.drop(columns = ["Add_1", "Add_2"], inplace = True)
df.head()
```

##### Approach 2 : applying lambda function to series

df["City"] = df[df["Purchase Address"].str.split(",")]

```python
sales_per_city = df.groupby("City").agg({"Sales": "sum"})
sales_per_city
```

#### Create visualization for the highest sales per city

```python
#cities = df["City"].unique()
cities = [i for i, df in df.groupby("City")]
plt.bar(cities, sales_per_city["Sales"])
plt.xticks(cities, rotation = "vertical", size = 12)
plt.xlabel("State")
plt.ylabel("Sales in USD ($)")
plt.title("Sales Per City For Year Ending 2019")
plt.style.use("ggplot")
plt.show()
```

#### What time should we display adverts to maximize likelihood of customers buying products.

```python
df.head()
```

##### Day Column 

```python
df["Date_Ordered"] = pd.to_datetime(df["Date_Ordered"])
df.head()
```

```python
df["Date_Ordered"].loc[0]
```

```python
df["Weekday"] = df["Date_Ordered"].dt.day_name()
df.head()
```

```python
df["Hour"] = df["Date_Ordered"].dt.hour
df["Minutes"] = df["Date_Ordered"].dt.minute
df.head()
```

```python
hourly_sales = [i for i, df in df.groupby("Hour")]
sales_per_hour = df.groupby("Hour").count()
plt.plot(hourly_sales, sales_per_hour["Sales"])
plt.xticks(hourly_sales, size = 13)
#plt.legend("Orders")
plt.xlabel("Hours", size = 13)
plt.ylabel("No. of Orders Recieved")
plt.title("Sales Per Hour", size = 14)
plt.style.use("ggplot")
plt.show()
```

##### Conclusion : From the Hours analysis, we can confidently suggest that the best time for advertising our products
##### range from 10 - 12 or 18 - 19, since those are the peak hours for recieving orders.

#### What Products are most often sold together

```python
product_sold = df.groupby("Product").sum()
product_sold[["Qty", "Price", "Sales"]]
```

##### Creating a subplot to visualize the above information

```python
products = [i for i, df in df.groupby("Product")]


# Visualization Args
barcolors = ['#87B88C', '#9ED2A1', '#E7E8CB', '#48A0C9', '#2A58A1', '#2E8B55', '#DF3659', 'Grey']
barstyle = {"edgecolor": "black", "linewidth": 1}

fig, ax = plt.subplots(3, 1, figsize = (15, 20),  gridspec_kw={'hspace':0.7}) #sharex = True


product_sold["Qty"].plot.bar(color = barcolors[6], **barstyle, ax = ax[0])
ax[0].set_title("Products Sold Per Quatity")
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation = 45)

product_sold["Price"].plot.bar(color = barcolors[7], **barstyle, ax = ax[1])
ax[1].set_title("Price Per Product")
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation = 45)

product_sold["Sales"].plot.bar(color = barcolors[5], **barstyle, ax = ax[2])
ax[2].set_title("Annual Sales Per Product")
ax[2].set_xticklabels(ax[2].get_xticklabels(), rotation = 45)

plt.show()
```
