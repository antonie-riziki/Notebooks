### EDA To Create a Classifier Model for Predicting Muffin or Cupcake Recipes
#### Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

from sklearn import svm

sb.set_style("darkgrid")
sb.set(font_scale = 1.2)

%matplotlib inline
```

#### Import Data

```python
df = pd.read_csv(r"D:\Open Classroom\Data Files\muffin-cupcake-master\muffin-cupcake-master\recipes_muffins_cupcakes.csv")
pd.set_option("display.max_rows", 100)
df.head(50)
```

#### Data Preparation

```python
sb.lmplot(x = "Sugar", y = "Butter", data = df, hue = "Type", fit_reg = True,
          palette = "Set1", scatter_kws = {"s": 70})
```

#### Fit the Model

#### Support Vector Machine
#### Is a model that best splits data of support vectors using the hyperplane

```python
sugar_butter = df[["Sugar", "Butter"]].to_numpy() # Input Variables
type_label = np.where(df["Type"] == "Muffin", 0, 1) #Prediction Model [Output]

model = svm.SVC(kernel = "linear")
model.fit(sugar_butter, type_label)

w = model.coef_[0]
a = -w[0] / w[1]
a
```
