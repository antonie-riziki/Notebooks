```python
import numpy as np

def get_response(data):
    # Get the similarity scores for the given data
    similarity_scores = sig[indices[data]]
    
    # Ensure similarity_scores is a 1D array
    if len(similarity_scores.shape) > 1:
        # Reduce to 1D using the mean
        similarity_scores = np.mean(similarity_scores, axis=1)
        
    # Enumerate and sort the similarity scores
    sorted_list = sorted(enumerate(similarity_scores), key=lambda x: x[1], reverse=True)
    
    # Get the top 10 indices based on similarity
    top_ten_indices = [x[0] for x in sorted_list[:10]]
    
    print('Recommended Response:')
    
    # Return the top 10 responses
    return new_df[['response']].iloc[top_ten_indices]

# Example usage
top_responses = get_response("I hate you")
print(top_responses)
```
