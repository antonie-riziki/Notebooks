```python
#Join - 
df1 = pd.DataFrame({"employee": ["Bob", "Jake", "Lisa", "Sue", "Leah"],
                   "department": ["Accounting", "Engineering", "Engineering", "HR", "Engineering"]})
df2 = pd.DataFrame({"employee": ["Bob", "Jake", "Lisa", "Sue"],
                   "date": [2004, 2008, 2012, 2014]})
```

```python
df1 
```

```python
df2
```

```python
#how = "left"
#gives access to records that dont match in both dataframes
df3 = pd.merge(df1, df2, how = "left")
df3
```

```python
#Assuming we have another table of supervisors
#Where each employee can have only 1 supervisor 
#But a supervisor can be allocated more than 1 employee
df4 = pd.DataFrame({"department": ["Accounting", "Engineering", "HR"], 
                   "supervisor": ["Carly", "Kurt", "Steve"]})
df4
```

```python
#Joining df3 and df4 
#One to many type of relationship
dfRecord = pd.merge(df3, df4)
dfRecord
```

```python
#Creating many to many relations in dataframes
df5 = pd.DataFrame({"department": ["Accounting", "Accounting", "Engineering", "Engineering", "HR", "HR"],
                   "competence": ["Math", "Spreadsheet", "Coding", "linux", "Spreadsheet", "Organization"]})
```

```python
pd.merge(dfRecord, df5)
```
