```python
df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
```

```python
company = df.groupby(df['Company'])
ticker = ['AAPL', 'AMZN', 'GOOGL', 'JP-MRGN', 'M-SOFT', 'NTFLX', 'NVDA','TSLA', 'VISA', 'WLMART']
for i in ticker:
    co = company.get_group(i)
    plt.figure(figsize = (15, 10))
    plt.plot(co['High'])
    plt.title('stock price for ' + i, fontdict={'size':15})
    plt.xticks(rotation = -45)
```
