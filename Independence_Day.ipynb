```python
#We take a quick look at csv module
#Using the csv context manager to acces, modifies and perform other operations on csv file(s)
#Accessing the 2018 csv file SOF Survey
with open("D:\Open Classroom\StackOverflow Dev Survey\developer_survey_2018\survey_results_public.csv", encoding = "utf-8") as file:
    fileReader = csv.DictReader(file)
    fileCounter = Counter()
    total = 0
    
    for line in fileReader:
        employment = line["Employment"].split(";")
        total += 1
        
        for emp in employment:
            fileCounter[emp] += 1
    for empStatus, val in fileCounter.most_common():
        empPercent = (val / total) * 100
        empPercent = round(empPercent, 2)
        print(f"{empStatus} : {empPercent}")        
```
